```python
import numpy as np
import pandas as pd
```

```python
def knapsack(items, max_capacity):
    a = np.ones(shape=(items.shape[0] + 1,max_capacity + 1))
    for i in range(a.shape[0]):
        for j in range(a.shape[1]):
            if i == 0 or j == 0:
                a[i][j] = 0
            else:
                if(items[i-1][0] > j):
                    a[i][j] = a[i-1][j]
                else:
                    a[i][j] = max(items[i-1][1] + a[i-1][j - items[i-1][0]],a[i-1][j]) 
    return(a)
```

```python
items = np.array([[4, 7], [5, 5], [6, 4], [8, 7]])
max_capacity = 13
a = knapsack(items, max_capacity)
print(a)
```
